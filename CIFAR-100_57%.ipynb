```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets.cifar100 import load_data

# CIFAR-100 데이터셋 로드
(train_images, train_labels), (test_images, test_labels) = load_data()

# 이미지 크기 정규화
train_images = train_images / 255.0
test_images = test_images / 255.0

# 레이블을 one-hot 인코딩으로 변환
train_labels_one_hot = tf.keras.utils.to_categorical(train_labels, 100)  # 100개의 클래스로 변환
test_labels_one_hot = tf.keras.utils.to_categorical(test_labels, 100)

# 첫 번째 10개 샘플 확인
def plot_sample_images(images, labels):
    images = images.reshape(-1, 32, 32, 3)  # 원래 형태로 복원
    plt.figure(figsize=(10, 5))
    for i in range(10):
        plt.subplot(2, 5, i + 1)
        plt.imshow(images[i])
        plt.title(labels[i][0])  # 클래스 번호만 표시
        plt.axis('off')
    plt.show()

# 첫 번째 10개 이미지 시각화
plot_sample_images(train_images, train_labels)

# 모델 정의
model = tf.keras.Sequential([
    # Conv1: 3x3 필터 32개, 활성화 함수 ReLU
    tf.keras.layers.Conv2D(32, (3, 3), padding='same', input_shape=(32, 32, 3)),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),  # 배치 정규화
    tf.keras.layers.Conv2D(32, (3, 3), padding='same'),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),  # 배치 정규화
    tf.keras.layers.MaxPooling2D((2, 2)),  # MaxPooling
    tf.keras.layers.Dropout(0.3),  # Dropout 비율 증가

    # Conv2: 3x3 필터 64개, 활성화 함수 ReLU
    tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(64, (3, 3), padding='same'),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.3),

    # Conv3: 3x3 필터 128개, 활성화 함수 ReLU
    tf.keras.layers.Conv2D(128, (3, 3), padding='same'),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Conv2D(128, (3, 3), padding='same'),
    tf.keras.layers.Activation('relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.3),

    # 특징을 추출한 뒤 Flatten
    tf.keras.layers.Flatten(),

    # Dense 층 (출력층 전에)
    tf.keras.layers.Dense(256, activation='relu'),  # 중간 Dense 층 추가
    tf.keras.layers.Dropout(0.5),  # Dropout 비율 증가

    # 출력층: 소프트맥스 활성화 (100개의 클래스)
    tf.keras.layers.Dense(100, activation='softmax')
])

# 모델 컴파일
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.00075),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# 모델 요약 출력
model.summary()

# 학습 및 평가
history = model.fit(train_images, train_labels_one_hot, epochs=50, batch_size=64, validation_data=(test_images, test_labels_one_hot))

# 테스트 결과 출력
test_loss, test_accuracy = model.evaluate(test_images, test_labels_one_hot, verbose=2)
print(f"Test Loss: {test_loss:.4f}, Test Accuracy: {test_accuracy:.4f}")

# 학습 그래프 출력
plt.figure(figsize=(12, 4))
plt.plot(history.history['loss'], 'b--', label='Loss')  # Loss는 파란색 점선
plt.plot(history.history['accuracy'], 'g-', label='Accuracy')  # Accuracy는 녹색 실선
plt.xlabel('Epoch')
plt.legend()
plt.show()

# 테스트 데이터에 대한 예측
predictions = model.predict(test_images[:15])

# 테스트 이미지 15개 시각화 함수 수정
def plot_test_images(images, predictions, labels_one_hot):
    plt.figure(figsize=(15, 10))
    for i in range(15):
        plt.subplot(3, 5, i + 1)
        plt.imshow(images[i])
        predicted_label = np.argmax(predictions[i])  # 예측된 클래스 번호
        plt.title(predicted_label)  # 예측 표시
        plt.axis('off')
    plt.show()

# 테스트 이미지 15개 출력
plot_test_images(test_images[:15], predictions, test_labels_one_hot[:15])
```
